# Variational circuit with amplitude embedding

```python
import pennylane as qml
from pennylane import numpy as np
import pennylane.optimize as optimize
from pennylane.optimize import AdamOptimizer
import pandas as pd

def Y_transform(Y):
    """Transforms a vector of 0s and 1s into a vector of -1s and 1s respectively.

    Args:
        - Y (list(int)): boolean input vector.  

    Returns:
        - (list(int)): transformed vector.
    """
    Y_t=[int(Y[i])*2-1 for i in range(len(Y))]
    return Y_t

def reverse_Y_transform(Y):
    """Transforms a vector of -1s and 1s into a vector of 0s and 1s respectively.

    Args:
        - Y (list(int)): input vector of 1s and -1s.  

    Returns:
        - (list(int)): transformed vector.
    """
    Y_t=[(1+Y[i])/2 for i in range(len(Y))]
    return Y_t

def square_loss(labels, predictions):
    """Computes the standard square loss between model predictions and true labels.

    Args:
        - labels (list(int)): True labels.
        - predictions (list(int)): Model predictions.

    Returns:
        - loss (float): the square loss.
    """

    loss = 0
    for l, p in zip(labels, predictions):
        loss = loss + (l - p) ** 2

    loss = loss / len(labels)
    return loss


def accuracy(labels, predictions):
    """Computes the accuracy of the model's predictions against the true labels.

    Args:
        - labels (list(int)): True labels.
        - predictions (list(int)): Model predictions.

    Returns:
        - acc (float): The accuracy.
    """

    acc = 0
    for l, p in zip(labels, predictions):
        if abs(l - p) < 1e-5:
            acc = acc + 1
    acc = acc / len(labels)

    return acc

def classify_data(X_train, Y_train, X_val, Y_val,num_layers):
    """classify data.

    Args:
        - X_train (np.ndarray): train data.
        - Y_train (np.ndarray): train labels.
        - X_val (np.ndarray): validation data.
        - Y_val (np.ndarray): validation labels.

    Returns:
        - predictions (list(int)): Your final model predictions on the train data.
        - predictions_val (list(int)): Your final model predictions on the validation data.
        - accs_train (list(int)): Accuracy of train predictions for each epoch.
        - accs_val (list(int)): Accuracy of validation predictions for each epoch.

    """
    
    num_wires = 2
    def layer(W):

        #rotations gates
        for i in range(num_wires):
            qml.RX(W[i,0],wires=i)
            qml.RZ(W[i,1],wires=i)
        
       #entangling gates
        for i in range(num_wires-1):
            qml.CNOT(wires=[i, i+1])
        qml.CNOT(wires=[num_wires-1, 0])
     
    
    #quantum embedding
    def statepreparation(x):
        qml.AmplitudeEmbedding(features=x, wires=range(num_wires),normalize=True)

    
    dev = qml.device("default.qubit", wires=num_wires) 

    # Defining a variational circuit
    @qml.qnode(dev)
    def circuit(weights,x):
        
        statepreparation(x)

        for W in weights:
            layer(W)
            
        return qml.expval(qml.PauliZ(0))
    
    
    # Defining a cost function 
    def cost(weights,X,Y):
   
        predictions = [circuit(weights,x) for x in X]
        return square_loss(Y, predictions)

    
    np.random.seed(0)
    
    
    weights_init = 0.01 * np.random.randn(num_layers, num_wires, 2, requires_grad=True)
    
    
    opt = AdamOptimizer(0.5)
    epochs = 25
   
    
    weights=weights_init
    accs_train=[]
    accs_val=[]
  
   
    for epoch in range(epochs):
        
        weights,_,_=opt.step(cost, weights, X_train, Y_train)
     
    
        predictions=[int(np.sign(circuit(weights,x))) for x in X_train]        
        acc=accuracy(Y_train,predictions)
        accs_train.append(acc)
        
        pred_val=[int(np.sign(circuit(weights,x))) for x in X_val] 
        acc_val=accuracy(Y_val,pred_val)
        accs_val.append(acc_val)
        
        #print(f"Epoch: {epoch}  | Accuracy_train: {acc} | Accuracy_val: {acc_val}")
       
        
    
    return predictions,pred_val,accs_train,accs_val

def plot_acc(train_acc,val_acc,epochs,num_layers):
    x=[i for i in range(epochs)]
    plt.plot(x, train_acc, '-')
    plt.plot(x, val_acc, '-')
    plt.legend(('train_accuracy', 'test_acuracy'),loc='upper right')
    plt.title(f'Accuracy with {num_layers} layers')
    plt.show()


if __name__ == "__main__":
    
    train=pd.read_csv("./mock_train_set.csv")
    test=pd.read_csv("./mock_test_set.csv")
    
    X_train=np.array(train.values[:,:4].tolist(),requires_grad=False)
    Y_train=np.array(Y_transform(train.values[:,4].tolist()),requires_grad=False)
    
    X_test=np.array(test.values[:,:4].tolist(),requires_grad=False)
    Y_test=np.array(Y_transform(test.values[:,4].tolist()),requires_grad=False)

    
    epochs=25
    num_layers=[2,3,4]
    for i in range(len(num_layers)):
        n=num_layers[i]
        predictions,predictions_val,acc_train,acc_test = classify_data(X_train, Y_train, X_test,Y_test,n)
        plt.subplot(3, 1, i+1)
        plot_acc(acc_train,acc_test,epochs,n)
    
    predictions=reverse_Y_transform(predictions)
    predictions_val=reverse_Y_transform(predictions_val)
    
```

With amplitude embedding and a variational circuit consisting of iterating over the layer with RX, RZ and CNOT gates, the best number of repetitions of this variational layer is 3. In fact, results are similar between the 3-layer and 4-layer variational circuits in terms of train accuracy meaning that starting from 3 layers, the expressibility of the above structure of the variational circuit is saturated. Whereas, for the 2-layer variational circuit, train accuracy is smaller that the best attained in the two following circuits(84%).

Note: The validation accuracy reaches in all cases 100% meaning that the validation data has a very particular distribution making it easy to predict.
